/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
query1 = 'SELECT name AS facility, SUM(CASE WHEN memid = 0 THEN guestcost ELSE membercost END ) AS revenue FROM Bookings LEFT JOIN Facilities USING ( facid ) GROUP BY name HAVING revenue <1000 ORDER BY revenue'

res = con.execute(query1)
df_q10 = pd.DataFrame(res)
df_q10.columns = res.keys()
df_q10
```

/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */


```python
query2 = 'SELECT m1.firstname, m1.surname, m2.firstname AS recommendation_firstname, m2.surname AS recommendation_surname FROM Members AS m1 LEFT JOIN Members AS m2 ON m1.recommendedby = m2.memid ORDER BY m1.surname, m1.firstname'

res = con.execute(query2)
df_q11 = pd.DataFrame(res)
df_q11.columns = res.keys()
df_q11
```

/* Q12: Find the facilities with their usage by member, but not guests */

```python
query3 = 'SELECT name AS facility, SUM(CASE WHEN b.memid = 0 THEN 0 ELSE 1 END) AS fac_usage FROM Facilities AS f INNER JOIN Bookings AS b USING (facid) GROUP BY facility'

res = con.execute(query3)
df_q12 = pd.DataFrame(res)
df_q12.columns = res.keys()
df_q12
```

/* Q13: Find the facilities usage by month, but not guests */

```python
query4 = "SELECT CASE WHEN starttime LIKE '2012-07-%' THEN '7' WHEN starttime LIKE '2012-08-%' THEN '8' ELSE '9' END AS month , name AS facility, SUM(CASE WHEN b.memid = 0.0 THEN 0 ELSE 1.0 END) AS monthly_usage FROM Bookings AS b INNER JOIN Facilities USING ( facid ) GROUP BY month , name"

res = con.execute(query4)
df_q13 = pd.DataFrame(res)
df_q13.columns = res.keys()
df_q13
```
